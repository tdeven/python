```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y,c=color,lw=4.0)
plt.show()
```

```python
#ColorMap-->A colormap is like a list of colors, where each colora has a value that ranges from 0 to 100.
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
color=np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y,c = color, cmap="viridis", lw= 4.0,s = sizes,alpha=0.5)#alpha-->Adjust the transparency of the dots.
plt.colorbar()#colormap in the drawing by including the plt.colorbar()statement.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size = (100))
y = np.random.randint(100, size = (100))
colors = np.random.randint(100,size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x,y, c = colors, s= sizes,alpha = 0.5,cmap="nipy_spectral")
plt.colorbar()
plt.show()
```
