```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,6])
y = np.array([0,250])

plt.plot(x,y,c="g")
plt.xlabel("Average Pulse")#xlabel() and ylabel()--> to set a label for the x-axis and y-axis
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,6])
y = np.array([0,250])

plt.plot(x,y,c="g")
plt.xlabel("Average Pulse")#xlabel() and ylabel()--> to set a label for the x-axis and y-axis
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")#title()-->to set a title for the plot.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,6])
y = np.array([0,250])

font1 = {'family':"serif","color":"blue","size":20}
font2= {'family': "serif", "color":"darkred", "size":15}

plt.plot(x,y,c="g")
plt.xlabel("Average Pulse", fontdict = font1)#xlabel() and ylabel()--> to set a label for the x-axis and y-axis
plt.ylabel("Calorie Burnage", fontdict = font2)#fontdict-->to set font properties for the title and labels.
plt.title("Sports Watch Data", fontdict= font2)#title()-->to set a title for the plot.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,6])
y = np.array([0,250])

#loc --> to set the position of title.
plt.plot(x,y,c="g",ls="-.", lw = 4.0)
plt.xlabel("Average Pulse")#xlabel() and ylabel()--> to set a label for the x-axis and y-axis
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data", loc = "center")#title()-->to set a title for the plot. values are 'right', 'left' and 'center'. default value is center

plt.show() 
```
