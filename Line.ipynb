```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = ([3,8,1,10])
plt.plot(ypoints, ls = "dashed")#linestyle =ls --> to change the style of the plotted line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = ([3,8,1,10])
plt.plot(ypoints,ls = "-.",color = "m")#color --> to set the color of the line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = ([3,8,1,10])
plt.plot(ypoints,ls = "solid",color = "blue", linewidth = "20.5")#color --> to set the color of the line
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3,8,1,10])
y2 = np.array([2,7,2,9])

plt.plot(y1,linestyle="-.",color="g",linewidth="4.0")
plt.plot(y2,ls = "solid", c = "b",lw = "3.0")

plt.show()
```
