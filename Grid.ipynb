```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sport Watch Data",size = 20, color = "blue")
plt.xlabel("Average Pulse",size=15)
plt.ylabel("Calorie Burnage",size=15)

plt.plot(x,y, c="g")
plt.grid()#grid()--> to add grid line  to the plot.
plt.show()
```
