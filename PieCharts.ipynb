```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas" , "Cherries" , "Dates"]

plt.pie(y,labels = mylabels) #label --> an array with one label for each wedge.
plt.show()
```

```python
#startangle-->With an angle in degrees, default angel is 0.
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels =["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels , startangle = 90)
plt.show()
```

```python
#explode--> must be an aray with one value for each wedge.
import matplotlib.pyplot as plt 
import numpy as np

y = np.array([35,25,25,15])
mylabels =["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [.20, 0.1, 0.3, 0.12] # Explode value must be less than one is better

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
#shadow --> by setting the shadows paramter to True.
import matplotlib.pyplot as plt 
import numpy as np

y = np.array([35,25,25,15])
mylabels =["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [.20, 0.1, 0.3, 0.12] # Explode value must be less than one is better

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```
